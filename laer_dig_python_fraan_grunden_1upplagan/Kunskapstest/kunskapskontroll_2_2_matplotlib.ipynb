### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: En placeholder för själva grundytan, som utan ytterligare definition blir tom
Axes: Plottningsområdet, dvs den yta det kommer att dras upp axlar och ritas på
Axis: Det man matematiskt kallar "axlar", t ex den skalade x-axeln i ett koordinatsystem
Artist: Alla delar som ritas/skrivs och är synliga är en "artist" t ex axlar, texter, punktdatan

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Med OO metod

X = np.linspace(-4, 4, num=80)
Y = X**2
fig, ax = plt.subplots()
ax.plot(X, Y)
ax.set_title("y=x^2 från -4 till 4")
ax.set_xlabel("x-axeln med 80 punkter mellan -4 och 4")
ax.set_ylabel("y-axeln visar x^2")
plt.show()
```

```python
# Med ej OO metod

X = np.linspace(-4, 4, num=80)
Y = X**2

plt.figure() # Ytan skapas
plt.title("y=x^2 från -4 till 4")
plt.xlabel("x-axeln med 80 punkter mellan -4 och 4")
plt.ylabel("y-axeln visar x^2")
plt.plot(X, Y)
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25} # dict data

# Datan delas upp
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(2, 1, figsize=(6, 9)) # 2 plottar på ytan i "2 rader 1 kolumn", vill ha en avlång canvas och sätter 8:12
plt.tight_layout(pad=3.0) # Skapa hspace mellan graferna

# Första plotten (index 0)
axs[0].set_aspect('equal') # Samma skala på axlarna

axs[0].set_xlim(-4, 4) # Luftigare plot
axs[0].set_ylim(-4, 4) # Luftigare plot

# Scatter-plot med x och y.  marker='.', s=3 för att göra lagom stor punkt
axs[0].scatter(random_data_x, random_data_y, marker='.', s=3) 
axs[0].set_title("Scatter plot of X and Y") # Titel på plot

# Andra plotten (index 1)
axs[1].bar(names, values) # bar chart med namn och värden
axs[1].set_title("Bar chart of X and Y") # Titel på plot

# Exekvera plotten
plt.show()
```
